To start our infinite row of vortices let's import our libraries and create our meshgrid.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

N = 200                        # Number of points in each direction (spatial resolution)
xStart,xEnd = -4.0,4.0         # X direction limits
yStart,yEnd = -2.0,2.0         # Y direction limits
x = np.linspace(xStart,xEnd,N) # X linear array of N elements
y = np.linspace(yStart,yEnd,N) # Y linear array of N elements
X,Y = np.meshgrid(x,y)         # Cartesian meshgrid
```

## Finite Row of Vortices

To make life easier, let's make functions to calculate our velocity field and stream function:

```python
def getvelocityvortex(strength,xv,yv,X,Y):
    u = strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = -strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v

def getstreamfunctionvortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi
```

Let's put in our parameters for our finite row of vortices:

```python
gamma = 5.0                           # Vortex strength
Nv = 10                               # Number of vortices
xVortex = np.linspace(xStart,xEnd,Nv) # X positions of vortices
yVortex = np.zeros(Nv)                # Y positions of vortices (chosen to be on X axis)
u = np.zeros((N,N))                   # Need to predefine our variable sizes to fit mesh
v = np.zeros((N,N))                   # But make them empty so our for loop can fill them
```

Now let's setup our for loop.  We want our for loop to use our two vortex position vectors above and calculate the velocity field of each.  We will then use our empty shell velocity vectors created above and use the for loop to sum our fields and populate the vector mesh.

```python
for i in range(0,Nv):                            
    uVortex,vVortex = getvelocityvortex(gamma,xVortex[i],yVortex[i],X,Y)
    psiVortex = getstreamfunctionvortex(gamma,xVortex[i],yVortex[i],X,Y)
    u = u + uVortex                   # You can see that our for loop is going through each vortex position from
    v = v + vVortex                   # our position vectors and storing the velocity field information in our mesh vectors
```

Now that our velocity field has been calculated, let's graph it.

```python
# No surprises here!
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Finite Vortices')
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortex,yVortex,c='#CD2305',s=80,marker='o')

plt.show()
```

# Infinite Row of Vortices

Let's take a look at what an infinite row of vortices looks like.  First we'll need to define some new variables:

```python
a = 1                                   # This is the separation of our vortices
xinfVortex = np.arange(xStart,xEnd,a)   # This is the new x position vector (no longer using np.linspace)
yinfVortex = np.zeros_like(xinfVortex)  # This is the new y position vector
```

We need to make a new function and calculate our new velocity fields.

```python
def getvelocityinfvortex(strength,xv,yv,X,Y):
    u = strength/(2*a)*np.sinh(2*pi*Y/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    v = -strength/(2*a)*np.sin(2*pi*X/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    return u,v

uinfVortex,vinfVortex = getvelocityinfvortex(gamma,xinfVortex,yinfVortex,X,Y)
```

Let's graph it using our Standard Operating Procedure (SOP) with a minor adjustment.

```python
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Infinite Vortices')
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uinfVortex,vinfVortex,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xinfVortex,yinfVortex,c='#CD2305',s=80,marker='o')  # Here we are graphing our position vectors that we made

plt.show()
```

Notice how the second graph takes vortices into account that couldn't have been considered in the first method we used.  We could have approximated an infinite row of vortices by extending the vortices we are calculating to very large numbers, but the for loop is a very inefficient way to process that information.  Let's see what happens, shall we?

```python
gamma = 5.0
Nv = 100
xVortex1 = np.linspace(-50,50,Nv)
yVortex1 = np.zeros(Nv)
u1 = np.zeros((N,N))
v1 = np.zeros((N,N))

for i in range(0,Nv):
    uVortex1,vVortex1 = getvelocityvortex(gamma,xVortex1[i],yVortex1[i],X,Y)
    psiVortex1 = getstreamfunctionvortex(gamma,xVortex1[i],yVortex1[i],X,Y)
    u1 = u1 + uVortex1
    v1 = v1 + vVortex1

size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Finite Vortices')
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u1,v1,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortex1,yVortex1,c='#CD2305',s=80,marker='o')


plt.show()
```
